```python
import numpy as np
import pandas as pd

data = {
    "days" : [1,2,3,4,5,6,7],
    "coins": [2,4,36,9,10,10,2]
}
df = pd.DataFrame(data)
print(df)
```

```python
import matplotlib.pyplot as plt
plt.scatter(df["days"],df["coins"], color="blue")
plt.title("Days vs Coins")
plt.xlabel("Days")
plt.ylabel("Coins")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
X = df[["days"]]
Y = df["coins"]
model.fit(X,Y)
print("W", model.coef_)
print("b", model.intercept_)
```

```python
# Prognoz - Predict
Y_pred = model.predict(X)

plt.scatter(X,Y, color="blue")
plt.plot(X, Y_pred, color="red")
plt.title("Days vs Coins")
plt.xlabel("Days")
plt.ylabel("Coins")
plt.show()
```

```python
new_x = [[10]]
predict_couns = model.predict(new_x)
print("Prediction: ", new_x[0][0], "will be",  predict_couns[0] )
```
