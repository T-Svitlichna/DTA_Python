### ✏️ Завдання 2.1

Є масив денних продажів магазину за 2 тижні (14 днів):

```python
sales = np.array([320, 415, 290, 480, 350, 620, 580,
                  310, 440, 305, 510, 370, 650, 595])
```

Виконай такі вибірки **без циклів**, лише через зрізи:
1. Продажі першого тижня (дні 1–7)
2. Продажі другого тижня (дні 8–14)
3. Продажі у вихідні: 6-й, 7-й, 13-й, 14-й день — використай індексацію списком `[[5,6,12,13]]`
4. Продажі через день, починаючи з першого (непарні дні: 1, 3, 5...)
5. Масив у зворотному порядку

````python
sales = np.array([320, 415, 290, 480, 350, 620, 580,
                  310, 440, 305, 510, 370, 650, 595])

# Твій код:
print("Продажі першого тижня (дні 1–7):", sales[:7].sum())
print("Продажі другого тижня (дні 8–14):", sales[7:14].sum())
print("Продажі у вихідні:", sales[[5,6,12,13]].sum())
print("Продажі через день, починаючи з першого (непарні дні: 1, 3, 5...):", sales[::2].sum())
print("Масив у зворотному порядку:", sales[::-1])
````

## Розділ 3 — Математичні операції та векторизація

> **Що повторюємо:** арифметика з масивами, `np.sqrt()`, `np.round()`,
> агрегатні функції `.sum()`, `.mean()`, `.std()`, `.min()`, `.max()`, `np.median()`

### ✏️ Завдання 3.1

Є масив цін товарів у доларах. Виконай **без циклів**:

```python
prices_usd = np.array([15.99, 42.50, 8.00, 125.00, 33.75, 67.20, 9.99])
```

1. Переведи ціни в гривні (курс: **1 USD = 41.5 UAH**), результат округли до 2 знаків
2. Нарахуй ПДВ 20% — помнож гривневі ціни на 1.2, округли до 2 знаків
3. Знайди: мінімальну, максимальну, середню ціну з ПДВ
4. Знайди суму всіх товарів з ПДВ — це загальна вартість кошика
5. Виведи результати з поясненнями

````python
prices_usd = np.array([15.99, 42.50, 8.00, 125.00, 33.75, 67.20, 9.99])

# Твій код:
print("Ціни у гривнях:", np.round(prices_usd * 41.5, 2))
print("Ціни з ПДВ:", np.round(prices_usd * 41.5 * 1.2, 2))
print("Мінімальна ціна з ПДВ:", np.round(prices_usd * 41.5 * 1.2, 2).min())
print("Максимальна ціна з ПДВ:", np.round(prices_usd * 41.5 * 1.2, 2).max())
print("Середня ціна з ПДВ:", np.round(prices_usd * 41.5 * 1.2 ).mean().round(2))
print("Сума всіх товарів з ПДВ:", np.round(prices_usd * 41.5 * 1.2, 2).sum())
````

### ✏️ Завдання 3.2

Є дані про кількість товарів на складі та їх собівартість.
Виконай **без циклів**:

```python
quantity  = np.array([120, 45, 300, 18, 75, 200, 60])
cost_each = np.array([250, 1800, 45, 9500, 380, 120, 750])
names     = ['Кабель', 'Клавіатура', 'Зошит', 'Монітор',
             'Мишка', 'Ручка', 'Навушники']
```

1. Порахуй загальну вартість кожного товару на складі (`quantity * cost_each`)
2. Знайди загальну вартість всього складу
3. Знайди товар з найбільшою сумарною вартістю — використай `np.argmax()`
4. Знайди товар з найменшою сумарною вартістю — `np.argmin()`
5. Виведи середню вартість одного найменування на складі

````python
quantity  = np.array([120, 45, 300, 18, 75, 200, 60])
cost_each = np.array([250, 1800, 45, 9500, 380, 120, 750])
names     = ['Кабель', 'Клавіатура', 'Зошит', 'Монітор',
             'Мишка', 'Ручка', 'Навушники']

# Твій код:
tot_sum = quantity * cost_each
print("Загальна вартість кожного товару на складі:", tot_sum)
print("Загальна вартість всього складу:", tot_sum.sum())
print("Товар з найбільшою сумарною вартістю:", names[tot_sum.argmax()])# імя/назва/значення - виведеного індексу
print("Товар з найменшою сумарною вартістю:", names[tot_sum.argmin()])
print("Середня вартість одного найменування на складі:", tot_sum.mean().round(2))
````

### ✏️ Завдання 4.1

Є масив часу виконання замовлень у хвилинах (норма — до 30 хв):

```python
delivery_time = np.array([22, 35, 28, 41, 19, 55, 27, 33, 24, 48,
                           31, 26, 52, 18, 29, 38, 23, 44, 20, 36])
```

1. Знайди всі замовлення, виконані **вчасно** (≤ 30 хв)
2. Знайди всі **запізнілі** замовлення (> 30 хв)
3. Порахуй кількість вчасних і запізнілих
4. Знайди відсоток запізнілих замовлень (округли до 1 знаку)
5. Яка середня затримка серед запізнілих замовлень?

````python
delivery_time = np.array([22, 35, 28, 41, 19, 55, 27, 33, 24, 48,
                           31, 26, 52, 18, 29, 38, 23, 44, 20, 36])

# Твій код:
mask = delivery_time <= 30
mask1 = delivery_time > 30
print("Вчасні замовлення:", delivery_time[mask])
print("Запізнілі замовлення:", delivery_time[mask1])
tot_order = delivery_time.size
print("Вчасні замовлення:", delivery_time[delivery_time <= 30].size)
print("Запізнілі замовлення:", delivery_time[delivery_time > 30].size)
print("Відсоток запізнілих замовлень:", round(delivery_time[mask1].size / tot_order * 100, 1), "%")
print("Середня затримка серед запізнілих замовлень:", round(delivery_time[mask1].mean(), 1))
````

1. Знайди дні з продажами **вище середнього** за місяць
