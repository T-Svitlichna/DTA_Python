# Розділ 1. Словники (`dict`)

**Швидке нагадування:**

Словник — це колекція пар `ключ : значення`, створюється у фігурних дужках:
```python
student = {"name": "Олена", "age": 20, "city": "Київ"}
```

**Основні операції:**
- Доступ: `student["name"]`
- Додавання/зміна: `student["email"] = "olena@example.com"`
- Видалення: `del student["age"]`
- Перевірка: `if "name" in student:`
- Безпечне отримання: `student.get("phone")` — повертає `None`, якщо ключа немає
- Довжина: `len(student)` — кількість пар

**Перебір у циклі:**
- `for key in d.keys():` — перебір ключів
- `for value in d.values():` — перебір значень
- `for key, value in d.items():` — перебір пар

### Завдання 22. Телефонна книга з пошуком ⭐⭐⭐

Створіть словник-телефонну книгу з 5 контактами. Напишіть програму, яка у циклі `while`:
1. Запитує у користувача дію: `пошук`, `додати`, `видалити`, `вихід`
2. Залежно від вибору — виконує відповідну дію
3. Завершується, коли користувач введе `вихід`

> 💡 **Підказка:** Використайте `while True:` (нескінченний цикл) і умову завершення з `break`, АБО змінну-прапорець, наприклад `running = True`.

````python
# Ваш код для завдання 22:
telefonna_kniga = {"Vasya": "+3508926644",
                   "Oksana":"+3508752158",
                   "Sveta": "+3507726644",
                   "Tommy": "+3508333338",
                   "Jeack": "+3507777774"}

while True:
    diya = input("Введіть дію (пошук / додати / видалити / вихід): ").strip().lower()

    if diya == "пошук":
        imya = input("Введіть ім'я для пошуку: ")
        if imya in telefonna_kniga:
            print(f"{imya}: {telefonna_kniga[imya]}")
        else:
            print("Контакт не знайдено")

    elif diya == "додати":
        imya = input("Введіть ім'я: ")
        nomer = int (input("Введіть номер: "))
        telefonna_kniga[imya] = nomer
        print(f"Контакт {imya} додано!")

    elif diya == "видалити":
        imya = input("Введіть ім'я для видалення: ")
        if imya in telefonna_kniga:
            del telefonna_kniga[imya]
            print(f"Контакт {imya} видалено!")
        else:
            print("Контакт не знайдено")

    elif diya == "вихід":
        print("До побачення!")
        break

    else:
        print("Невідома дія. Спробуйте ще раз.")
````

````python
# Ваш код для завдання 26:
history = []

while True:
    operation = input("Введіть операцію (+ - * / або вихід): ")

    if operation == "вихід":
        break

    if operation not in ["+", "-", "*", "/"]:
        print("Невідома операція, спробуй ще раз")
        continue

    a = float(input("Введіть число 1: "))
    b = float(input("Введіть число 2: "))

    if operation == "+":
        result = a + b
    elif operation == "-":
        result = a - b
    elif operation == "*":
        result = a * b
    elif operation == "/":
        if b == 0:
            print("Ділення на нуль!")
            continue
        result = a / b

    entry = f"{a} {operation} {b} = {result}" # заносить все в рядок
    history.append(entry)
    print("Результат:", result)

print("\n=== Історія обчислень ===")
for item in history:
    print(item)
````

````python
matches = [
    {"team_a": "Динамо",     "team_b": "Шахтар",     "score_a": 1, "score_b": 2},
    {"team_a": "Зоря",       "team_b": "Динамо",     "score_a": 0, "score_b": 1},
    {"team_a": "Шахтар",     "team_b": "Дніпро",     "score_a": 3, "score_b": 1},
    {"team_a": "Динамо",     "team_b": "Ворскла",    "score_a": 2, "score_b": 2},
    {"team_a": "Шахтар",     "team_b": "Зоря",       "score_a": 4, "score_b": 0},
    {"team_a": "Дніпро",     "team_b": "Ворскла",    "score_a": 1, "score_b": 1},
    {"team_a": "Зоря",       "team_b": "Дніпро",     "score_a": 2, "score_b": 1},
    {"team_a": "Ворскла",    "team_b": "Шахтар",     "score_a": 0, "score_b": 3}
]
# Ваш код:
points = {}

for match in matches:
    a = match["team_a"]
    b = match["team_b"]
    sa = match["score_a"]
    sb = match["score_b"]


    if a not in points:
        points[a] = 0
    if b not in points:
        points[b] = 0


    if sa > sb:
        points[a] += 3
    elif sa < sb:
        points[b] += 3
    else:
        points[a] += 1
        points[b] += 1


sorted_points = sorted(points.items(), key=lambda x: x[1], reverse=True)

print(f"{'Команда':10} | Очки")
print("-" * 20)
for team, pts in sorted_points:
    print(f"{team:10} | {pts}")

````

- `баланс` — показати баланс
- `зняти` — запитати суму і зняти (якщо вистачає коштів)
- `поповнити` — запитати суму і додати до балансу
- `історія` — показати список усіх операцій
- `вихід` — завершити

Зберігайте кожну операцію у списку як словник `{"type": "зняти", "amount": 500}`.

> 💡 **Підказка:** Створіть `history = []` перед циклом. Кожну операцію додавайте через `.append()`.

````python
# Ваш код для завдання 45:
history = []
balance = 5000
while True:
    diya = input("\nДія (баланс / зняти / поповнити / історія / вихід): ").strip().lower()

    if diya == "баланс":
        print("Баланс:", balance)
    elif diya == "зняти":
        amount = int(input("Введіть суму для зняття: "))
        if amount <= balance:
            balance -= amount
            history.append
        if amount > balance:
            print("Недостатньо коштів на рахунку.")
        elif amount <= 0:
            print("Сума некоректна, спробуйте ще раз пізніше")
    elif diya == "поповнити":
        amount = int(input("Введіть суму для поповнення: "))
        balance += amount
        history.append({"type": "поповнення", "amount": amount})
    elif diya == "історія":
        print("Історія операцій:")
        for oper in history:
            print(f"{operation['type']}: {operation['amount']}")
    elif diya == "вихід":
        print(f" До побачення! Ваш баланс: {balance} грн")
        break

    else:
        print(" Невідома дія. Спробуйте ще раз.")
````
